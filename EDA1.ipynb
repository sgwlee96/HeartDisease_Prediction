```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataRootPath = "../Machine_learning/data/"
```

```python
df = pd.read_csv(dataRootPath + "initial_data/frmgham2.csv")
print(df.shape)
df.head()
```

## EDA

```python
df.columns
```

```python
df.nunique().to_frame().reset_index().rename(columns={"index":"features", 0:"uniqueValue"})
```

```python
# find missing values for each column

missingValue = df.isna().sum().to_frame().reset_index().rename(columns={"index":"features", 0:"cnt_missingValues"})
missingValue = missingValue[missingValue["cnt_missingValues"]!= 0]   
missingValue
```

```python
print("HDLC column missing value percentage:" ,list(missingValue[missingValue['features']=='HDLC'].cnt_missingValues/df.shape[0])[0])
print("LDLC column missing value percentage:" ,list(missingValue[missingValue['features']=='LDLC'].cnt_missingValues/df.shape[0])[0])
```
